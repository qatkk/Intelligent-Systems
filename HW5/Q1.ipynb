```python
import numpy as np
import pandas as pd 
from sklearn.utils import shuffle
```

```python
trainData = pd.read_csv("Mushroom_Train.csv")
headers = trainData.columns
data = trainData.to_numpy()
data = shuffle(data)
label = data[:, 0]
x = data[:, 1:len(headers)]
print(f'number of features are {len(x[1])}')        
labels = np.zeros(len(label))
labels[np.where(label == 'p')[0]] = 1
```

```python
# Calculate prior probabilities 
# Being poisoness 
probP = sum(labels)/ len(labels)
probE = 1 - probP 
print(f'probability of being poisoness : {probP} and the probability of being edible is : {probE}')
probsPois = {}
probsEdible = {}
Values = {}           
# Calculate conditional probabilities 
for i in range(len(x[1])): 
  Values[i] = np.unique(x[:, i])
  temp = x[:, i]
  probsPois[i] = {}
  probsEdible[i] = {}
  for j in range(len(Values[i])):
    number = len(np.where(temp[np.where(labels == 1)[0]] == Values[i][j])[0])
    probsPois[i][j] = (number+1)/len(temp[np.where(labels ==0)[0]])
    number = len(np.where(temp[np.where(labels == 0)[0]] == Values[i][j])[0])
    probsEdible[i][j] = (number+1)/len(temp[np.where(labels == 1 )[0]])
    
print(f'Conditional probabilities of being poisoness {probsPois}')
print(f'Conditional probabilities of being edible {probsEdible}')
```

```python
testData = pd.read_csv("Mushroom_Test.csv")
headers = testData.columns
data = testData.to_numpy()
data = shuffle(data)
testLabel = data[:, 0]
testLabels = np.zeros(len(testLabel))
testLabels[np.where(testLabel == 'p')[0]] = 1
xTest = data[:, 1:len(headers)]
print(len(xTest[1]))
```

```python
# Making predictions
predictedLabels = np.zeros(1)
for i in range(len(xTest[:, 1])):
  probs = [probE, probP]
  for j in range(len(xTest[1, :])):
    attribute = int(np.where(Values[j] == xTest[i,j])[0])
    probs[0] = probs[0] * probsEdible[j][attribute]
    probs[1] = probs[1] * probsPois[j][attribute]
  predictedLabels = np.append(predictedLabels, np.argmax(probs))

predictedLabels = predictedLabels[1: len(predictedLabels)]
tp = len(np.where(testLabels[np.where(predictedLabels == 1)[0]]== 1)[0])
tn = len(np.where(testLabels[np.where(predictedLabels == 0)[0]]== 0)[0])
fp = len(np.where(testLabels[np.where(predictedLabels == 1)[0]]== 0)[0])
fn = len(np.where(testLabels[np.where(predictedLabels == 0)[0]]== 1)[0])
print(f'the confustion matrix will be as follows \n {tp, fp} \n {fn, tn}')
print(f'the accuracy of this method is {(tp + tn) / len(testLabels)}')
```
